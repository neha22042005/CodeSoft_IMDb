```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = pd.read_csv("IMDb Movies India.csv", encoding='latin1')
print(data.head())
```

```python
print(data.columns)
```

```python
data = data[['Genre', 'Director', 'Actor 1', 'Rating']]
print(data)


X = data[['Genre', 'Director', 'Actor 1']]
y = data['Rating']
```

```python
data = data.dropna()
print(data)
```

```python
data['Genre'] = data['Genre'].astype('category').cat.codes
data['Director'] = data['Director'].astype('category').cat.codes
data['Actor 1'] = data['Actor 1'].astype('category').cat.codes
print(data['Genre'])
print(data['Director'])
print(data['Actor 1'])
```

```python
X = data[['Genre', 'Director', 'Actor 1']]
y = data['Rating']
print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
print(model)
```

```python
score = model.score(X_test, y_test)
print("Model Score:", score)
```

```python
sample = pd.DataFrame([[1, 2, 3]], columns=['Genre', 'Director', 'Actor 1'])
prediction = model.predict(sample)

print("Predicted Rating:", prediction[0])
```
